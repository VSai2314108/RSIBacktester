```python
# Initialize SparkSession (if not already initialized)
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ReadPartitionedData").getOrCreate()

from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Define the schema including partition columns
schema = StructType([
    StructField("year", IntegerType(), True),
    StructField("month", IntegerType(), True),
    StructField("branch", StringType(), True),
    # Add other fields from your data here
])

df = spark.read.schema(schema).parquet('./output_data_spark') \
    .where(
        (F.col('year') == 2023) &
        (F.col('month') == 10) &
        (F.col('branch') == 'rsi-2-ERX-%3C-1-1-ERX')
    )
```
